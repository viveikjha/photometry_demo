```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python

def kewley01(log_nii_ha):
    """Kewley et al. (2001) maximum starburst line (extreme starburst classification line)"""
    return 0.61 / (log_nii_ha - 0.47) + 1.19

def kauffmann03(log_nii_ha):
    """Kauffmann et al. (2003) pure star-forming classification line"""
    return 0.61 / (log_nii_ha - 0.05) + 1.3
```

```python


# Define the x-axis range for the dividing lines
x_kauffmann = np.linspace(-2.0, 0.0, 1000)  # Kauffmann line valid for log([NII]/Ha) < 0.05
x_kauffmann = x_kauffmann[x_kauffmann < 0.05]

x_kewley = np.linspace(-2.0, 0.4, 1000)  # Kewley line valid for log([NII]/Ha) < 0.47
x_kewley = x_kewley[x_kewley < 0.47]

# Compute y values for the classification lines
y_kauffmann = kauffmann03(x_kauffmann)
y_kewley = kewley01(x_kewley)

# Seyfert/LINER dividing line (Schawinski et al. 2007)
# Vertical line at log([NII]/Ha) ~ 0.05 going from Kewley line upward
# Or a slanted line: log([OIII]/Hb) = 1.89 * log([NII]/Ha) + 0.76
x_seyfert_liner = np.linspace(-0.18, 0.6, 500)
y_seyfert_liner = 1.89 * x_seyfert_liner + 0.76

# Generate synthetic data for illustration
np.random.seed(42)

# Star-forming galaxies (below Kauffmann line)
n_sf = 500
sf_x = np.random.normal(-0.7, 0.3, n_sf)
sf_y = np.random.normal(-0.3, 0.3, n_sf)
# Keep only those below the Kauffmann line
mask_sf = sf_y < kauffmann03(sf_x)
sf_x, sf_y = sf_x[mask_sf], sf_y[mask_sf]

# Composite galaxies (between Kauffmann and Kewley lines)
n_comp = 300
comp_x = np.random.normal(-0.15, 0.15, n_comp)
comp_y = np.random.normal(0.1, 0.25, n_comp)
mask_comp = (comp_y > kauffmann03(comp_x)) & (comp_y < kewley01(comp_x))
comp_x, comp_y = comp_x[mask_comp], comp_y[mask_comp]

# Seyfert galaxies (above Kewley line, above Seyfert/LINER line)
n_sy = 200
sy_x = np.random.normal(0.0, 0.2, n_sy)
sy_y = np.random.normal(0.9, 0.2, n_sy)
mask_sy = (sy_y > kewley01(np.clip(sy_x, -2, 0.46))) & (sy_y > 1.89 * sy_x + 0.76)
sy_x, sy_y = sy_x[mask_sy], sy_y[mask_sy]

# LINER/AGN galaxies (above Kewley line, below Seyfert/LINER line)
n_agn = 200
agn_x = np.random.normal(0.2, 0.15, n_agn)
agn_y = np.random.normal(0.2, 0.2, n_agn)
mask_agn = (agn_y > kewley01(np.clip(agn_x, -2, 0.46))) & (agn_y < 1.89 * agn_x + 0.76)
agn_x, agn_y = agn_x[mask_agn], agn_y[mask_agn]
```

```python

# --- Plotting ---
fig, ax = plt.subplots(figsize=(10, 8))

# Plot the data points
ax.scatter(sf_x, sf_y, c='dodgerblue', s=20, alpha=0.6, label='Star-Forming', zorder=3)
ax.scatter(comp_x, comp_y, c='limegreen', s=20, alpha=0.6, label='Composite', zorder=3)
ax.scatter(sy_x, sy_y, c='red', s=20, alpha=0.6, label='Seyfert', zorder=3)
ax.scatter(agn_x, agn_y, c='orange', s=20, alpha=0.6, label='LINER/AGN', zorder=3)

# Plot the classification lines
ax.plot(x_kauffmann, y_kauffmann, 'k-', linewidth=2.5, label='Kauffmann+03', zorder=5)
ax.plot(x_kewley, y_kewley, 'k--', linewidth=2.5, label='Kewley+01', zorder=5)
ax.plot(x_seyfert_liner, y_seyfert_liner, 'k:', linewidth=2.5, label='Seyfert/LINER (Schawinski+07)', zorder=5)

# Add region labels
ax.text(-1.1, -0.8, 'Star-Forming', fontsize=16, fontweight='bold', color='dodgerblue',
        ha='center', va='center',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='dodgerblue', alpha=0.8))

ax.text(-0.15, 0.3, 'Composite', fontsize=13, fontweight='bold', color='green',
        ha='center', va='center',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='green', alpha=0.8))

ax.text(-0.3, 1.1, 'Seyfert', fontsize=16, fontweight='bold', color='red',
        ha='center', va='center',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='red', alpha=0.8))

ax.text(0.4, 0.0, 'LINER', fontsize=16, fontweight='bold', color='darkorange',
        ha='center', va='center',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='darkorange', alpha=0.8))

# Fill regions with transparent colors
# Star-forming region (below Kauffmann)
x_fill_sf = np.linspace(-2.0, 0.04, 500)
y_fill_sf_upper = kauffmann03(x_fill_sf)
ax.fill_between(x_fill_sf, -1.5, y_fill_sf_upper, alpha=0.07, color='blue', zorder=1)

# Composite region (between Kauffmann and Kewley)
x_fill_comp = np.linspace(-2.0, 0.04, 500)
y_fill_comp_lower = kauffmann03(x_fill_comp)
y_fill_comp_upper = kewley01(x_fill_comp)
ax.fill_between(x_fill_comp, y_fill_comp_lower, y_fill_comp_upper, alpha=0.07, color='green', zorder=1)

# Seyfert region (above Kewley line and above Seyfert/LINER line)
x_fill_agn = np.linspace(-0.18, 0.46, 300)
y_fill_agn_lower = kewley01(x_fill_agn)
y_fill_seyfert_liner = 1.89 * x_fill_agn + 0.76
# Upper part -> Seyfert
ax.fill_between(x_fill_agn, np.maximum(y_fill_agn_lower, y_fill_seyfert_liner), 1.5,
                alpha=0.07, color='red', zorder=1)
# Lower part -> LINER
ax.fill_between(x_fill_agn, y_fill_agn_lower, np.minimum(y_fill_seyfert_liner, 1.5),
                where=y_fill_seyfert_liner > y_fill_agn_lower, alpha=0.07, color='orange', zorder=1)

# Axis settings
ax.set_xlim(-2.0, 1.0)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel(r'log([NII] $\lambda$6583 / H$\alpha$)', fontsize=14)
ax.set_ylabel(r'log([OIII] $\lambda$5007 / H$\beta$)', fontsize=14)
ax.set_title('BPT Diagram — Galaxy Classification', fontsize=18, fontweight='bold')
ax.legend(loc='lower left', fontsize=10, framealpha=0.9)
ax.tick_params(axis='both', labelsize=12)
ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('bpt_diagram.png', dpi=200, bbox_inches='tight')
plt.show()
```
